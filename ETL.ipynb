### Primero realizo mi ETL en `movies_dataset`

Relleno los valores nulos de budget y revenue por `0`. Como Budget no tiene valores nulos, voy a dejarlo como estÃ¡.

```python
df_movies['revenue']=df_movies['revenue'].fillna(0)
```

Elimino los valores nulos del campo release_date

```python
df_movies=df_movies.dropna(subset=['release_date'])
```

Transformo el tipo de datos en la columna "release_date" a fecha

```python
df_movies['release_date']=pd.to_datetime(df_movies['release_date'],format='%Y-%m-%d',errors='coerce')
```

Verifico si tengo valores nulos

```python
df_movies['release_date'].isna().sum()
```

Veo cuales son esos valores nulos

```python
df_movies[df_movies['release_date'].isna()]
```

Elimino los valores nulos restantes de 'release_date'

```python
df_movies=df_movies.dropna(subset=['release_date'])
```

Verifico nuevamente los valores nulos

```python
df_movies['release_date'].isna().sum()
```

Creo una nueva columna 'release_year'extrayendo la informacion de 'release_date'

```python
df_movies['release_year']=df_movies['release_date'].dt.year
```

Creo la columna de retorno de inversion 'return', dividiendo los valores de la columna 'revenue' por los de la columna 'budget'.

Para realizar esto, debo de verificar si es posible matematicamente.

```python
df_movies.info()
```

Dado que la columna 'budget' esta siendo tratada como un string, procedo a modificar su tipo por float y asi corresponder con el tipo de la columna 'revenue'

```python
df_movies['budget']=pd.to_numeric(df_movies['budget'],downcast='float',errors='coerce')
```

Realizo la division y creo la columna 'return' en el proceso.

```python
df_movies['return']=df_movies['revenue'].divide(df_movies['budget']).fillna(0)
```

Algunos de los valores de la columna 'return' tienden a infinito, por lo cual voy a reemplazar esos valores por NaN

```python
df_movies['return'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

Relleno los NaN de 'return' por `0`

```python
df_movies['return'].fillna(0,inplace=True)
```

Procedo a desanidar los datos que pertenecen a ciertas columnas en forma de listas y diccionarios, para crear nuevas columnas y aÃ±adirlas al dataframe.

Inicio por la columna `belongs_to_colecction`

Relleno los valores nulos por llaves '{}', ya que de esta manera puedo usar la funcion map (eval) para reconocer que los datos son diccionarios.

```python
df_movies['belongs_to_collection']=df_movies['belongs_to_collection'].fillna('{}')
```

Reseteo los indices, ya que elimine previamente algunas filas

```python
df_movies = df_movies.reset_index(drop=True)
```

Utilizo json:normalize para obtener un dataframe con los datos en columnas individuales.

```python
df1=pd.json_normalize(df_movies['belongs_to_collection'].map(eval))
```

```python
df1.head()
```

Creo la columna 'collection_name' con la informacion de 'name'

```python
df_movies['collection_name']=df1['name']
```

Creo 'idcollection' con la informacion de 'id'

```python
df_movies['idcollection']=df1['id']
```

Elimino la columna 'belongs_to_collection'

```python
df_movies.drop(columns=['belongs_to_collection'],inplace=True)
```

Elimino las columnas que no voy a utilizar (video, imdb_id, adult, original_title, poster_path y homepage).

Como solo pase la informacion de las columnas de 'collection_name' y 'idcollection', ya quedo eliminada la columna 'poster_path'.

```python
df_movies.drop(columns=['video','imdb_id','adult','original_title','poster_path','homepage'], inplace=True)
```

```python
df_movies.head()
```

Elimino el dataframe adicional df1 que cree, ya que no lo necesito.

```python
del df1
```

Voy a transformar la columna `genres` en dos columnas, una con 'idgenres' y otra con 'genres_name'

Primero, voy a crear dos funciones (get_names y get_id) para extraer los nombres e id como lista.

```python
get_names = lambda lst: [d['name'] for d in lst]
get_id = lambda lst: [d['id'] for d in lst]
```

Transformo

```python
df_movies['genres']=df_movies['genres'].map(eval)
df_movies['genres_name'] = df_movies['genres'].apply(get_names)
df_movies['idgenres'] = df_movies['genres'].apply(get_id)
```

Elimino la columna 'genres'

```python
df_movies.drop(columns='genres',inplace=True)
df_movies.head()
```

Procedo a transformar la informacion de mi columna, `spoken_languages`.

Creo la funcion get_slcodes, la cual me va a devolver una lista de codigos de las lenguas habladas.

```python
get_slcodes = lambda lst: [d['iso_639_1'] for d in lst]
```

```python
df_movies['spoken_languages']=df_movies['spoken_languages'].map(eval)
```

Creo dos nuevas columnas, 'slang_name' para el nombre de la lengua y 'slang_isocode' para el iso del idioma.

```python
df_movies['slang_name'] = df_movies['spoken_languages'].apply(get_names)
df_movies['slang_isocode'] = df_movies['spoken_languages'].apply(get_slcodes)
```

Elimino la columna 'spoken_languages'

```python
df_movies.drop(columns='spoken_languages',inplace=True)
df_movies.head()
```

Procedo a transformar la columna `production_companies`

```python
df_movies['production_companies']=df_movies['production_companies'].map(eval)
```

Creo las columnas 'pcompany_name' y 'idpcompany', reutilizando mis funciones previas (get_names y get_id)

```python
df_movies['pcompany_name'] = df_movies['production_companies'].apply(get_names)
df_movies['idpcompany'] = df_movies['production_companies'].apply(get_id)
```

Elimino la columna 'production_companies'

```python
df_movies.drop(columns='production_companies',inplace=True)
df_movies.head()
```

Finalmente, voy a trabajar en `production_countries`.

Voy a crear una nueva funcion que obtenga las claves "iso_3166_1"

```python
get_pccodes = lambda lst: [d['iso_3166_1'] for d in lst]
```

```python
df_movies['production_countries']=df_movies['production_countries'].map(eval)
```

```python
df_movies['pcountry_name'] = df_movies['production_countries'].apply(get_names)
df_movies['pcountry_isocode'] = df_movies['production_countries'].apply(get_pccodes)
```

Elimino la columna 'production_countries'

```python
df_movies.drop(columns='production_countries',inplace=True)
df_movies.head()
```

Imputo los valores nulos de mi campo 'runtime' con la media de los valores

```python
median_runtime = df_movies['runtime'].median()
df_movies['runtime'] = df_movies['runtime'].fillna(median_runtime)
```

Elimino los valores nulos de 'original_language'

```python
df_movies = df_movies.dropna(subset=['original_language'])
```

Corrijo los tipos de mis campos

```python
df_movies['id'] = df_movies['id'].astype(int)
df_movies['runtime'] = df_movies['runtime'].astype(int)
```

```python
df_movies.isna().sum()
```

Reutilizo mis funciones get_names y get_id para transformar la columna 'cast'

```python
#Transformo la informacion de mi campo 'cast'
df_credits['cast']=df_credits['cast'].map(eval)
df_credits['namecast'] = df_credits['cast'].apply(get_names)
df_credits['idcast'] = df_credits['cast'].apply(get_id)
```

```python
#Elimino la columna 'cast'
df_credits.drop(columns='cast',inplace=True)
df_credits.head()
```

Finalmente transformo la columna 'crew'

```python
df_credits['crew']=df_credits['crew'].apply(eval)
```

```python
#funcion para extraer los campos director y name
def obtener_director(crew):
    directors = []
    for member in crew:
        if member['job'] == 'Director':
            directors.append(member['name'])
    return directors
```

```python
#aplicar la funcion a la columna crew y guardar los resultados en una nueva columna directors
df_credits['directors']=df_credits['crew'].apply(obtener_director)
```

```python
df_credits.drop(columns='crew',inplace=True)
df_credits.head()
```
